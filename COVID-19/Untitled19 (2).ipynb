```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
%matplotlib inline

pd.set_option('display.max_rows', None)
pd.options.display.float_format = '{:.2f}'.format

fig, ax = plt.subplots()
ax.xaxis.set_major_formatter(ScalarFormatter(useOffset=False))
ax.yaxis.set_major_formatter(ScalarFormatter(useOffset=False))

infect_global = pd.read_csv('感染人数.csv')  # 直接读取 CSV 文件并转换为 DataFrame
display(infect_global)
```

```python
infect_global.drop_duplicates(inplace=True)   # 执行去重
display(infect_global)
```

```python
display(infect_global.isnull().sum())          # 查询NaN空值
```

```python
delete_data=infect_global.fillna(0)        # 用0填充NaN
display(delete_data)
```

```python
display(delete_data.isnull().sum())          # 再次查询NaN空值
```

```python
infect_head=delete_data[1:]                                    # 获取除了第一行后面得数据
infect_head
```

```python
infect_region_vaccin=infect_head[["地区区域","疫苗接种剂量",'接种第一剂的人数','最后一剂接种人数']]    # 获取需要得三列
infect_region_vaccin
```

```python
infect_region_infect=infect_head[['地区区域','感染人数','死亡人数']]    # 获取需要得三列
infect_region_infect
```

```python
infect_region_group_first=infect_region_vaccin.groupby('地区区域')[['接种第一剂的人数','最后一剂接种人数']].sum() # 按区域进行人数汇总
infect_region_group_first=pd.DataFrame(infect_region_group_first)   # 把修改好得数据重新创建DataFrame类型
infect_region_group_first['地区区域数列']=infect_region_group_first.index   # 把索引得内容转为列
infect_region_group_first
```

```python
plt.title('全世界各区域感染人数和死亡人数对比',fontsize=20)
x=infect_region_infect['地区区域']
y1=infect_region_infect['感染人数']
y2=infect_region_infect['死亡人数']
plt.bar(x,y1,facecolor='lightskyblue')
plt.bar(x,y2,facecolor='red')
plt.legend(infect_region_infect.columns[1:3], loc='best',fontsize=12)
plt.show()                               # 按区域看出死亡人数和感染人数的对比

```

```python
plt.title('全世界各区域接种第一针和最后一帧的人数对比',fontsize=20)
x=infect_region_group_first['地区区域数列']
y1=infect_region_group_first['接种第一剂的人数']
y2=infect_region_group_first['最后一剂接种人数']

plt.bar(x,y1,facecolor='lightskyblue')   
plt.bar(x,y2,facecolor='red')
plt.legend(infect_region_group_first.columns, loc='best',fontsize=12)
plt.show()                               # 按区域看出死亡人数和感染人数的对比
```

```python
infect_region_infect_groupby=infect_region_infect.groupby('地区区域')['感染人数'].sum()
infect_region_groupby_data=pd.DataFrame(infect_region_infect_groupby)
infect_region_groupby_data['地区区域数列']=infect_region_groupby_data.index   # 把索引得内容转为列
infect_region_groupby_data
```

```python
labels = infect_region_groupby_data['地区区域数列']
data = infect_region_groupby_data['感染人数']

def plot_pie_charts(labels, y, title):
    c = ['lightcoral', 'rosybrown', 'sandybrown', 'navajowhite', 'gold', 'khaki', 'lightskyblue']
    plt.figure(figsize=(10, 8))
    plt.title(title, size=20)
    plt.pie(y, labels=labels, colors=c, autopct='%.1f%%')
    plt.legend(labels, loc='best', fontsize=12)
    plt.show()

plot_pie_charts(labels, data, '全世界各区域的感染人数比例')
```
